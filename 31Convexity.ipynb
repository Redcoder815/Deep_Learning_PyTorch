```
import matplotlib.pyplot as plt
import numpy as np
import torch
from mpl_toolkits import mplot3d
```

```
f = lambda x: 0.5 * x**2  # Convex
g = lambda x: torch.cos(np.pi * x)  # Nonconvex
h = lambda x: torch.exp(0.5 * x)  # Convex

x, segment = torch.arange(-2, 2, 0.01), torch.tensor([-1.5, 1])

_, axes = plt.subplots(1, 3, figsize=(9, 3))
for ax, func in zip(axes, [f, g, h]):
    # Plot the function curve
    ax.plot(x.detach().numpy(), func(x).detach().numpy())
    # Plot the line segment (representing convexity test)
    ax.plot(segment.detach().numpy(), func(segment).detach().numpy(), 'ro--')
    ax.set_title(func.__name__ if func.__name__ != '<lambda>' else 'Function')
    ax.grid(True)
plt.tight_layout()
plt.show()
```

```
f = lambda x: (x - 1) ** 2

plt.figure(figsize=(4.5, 3))
plt.plot(x.detach().numpy(), f(x).detach().numpy(), label='f(x)')
plt.plot(segment.detach().numpy(), f(segment).detach().numpy(), 'ro--', label='segment')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Function and Segment')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
